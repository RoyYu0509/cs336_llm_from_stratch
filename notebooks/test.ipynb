```python
import regex as re
def convert_2_bytes(text: str):
    """
    Convert the a sequence of text to a seqeunce of bytes
    """
    bytes_seq = []
    PAT = r"""'(?:[sdmt]|ll|ve|re)| ?\p{L}+| ?\p{N}+| ?[^\s\p{L}\p{N}]+|\s+(?!\S)|\s+"""
    for match in re.finditer(PAT, text):
        substr_enc = match.group().encode("utf-8")  # 
        bytes_seq += [bytes([b]) for b in substr_enc]
    return bytes_seq
```

```python
from collections.abc import Callable, Iterable
from typing import Optional
import torch
import math

class AdamW(torch.optim.Optimizer):
    def __init__(
            self, params, lr=1e-3, 
            b1=0.9, b2=0.95, 
            lbd=0.9, 
            eps=1e-8
        ):
        if lr < 0:
            raise ValueError(f"Invalid learning rate: {lr}")
        # Create the default hyperparamters value for different para_group
        defaults = {
            "lr": lr,
            "lambda": lbd,
            "eps": eps,
            "b1": b1,
            "b2": b2

        }
        super().__init__(params, defaults)

    def step(self, closure: Optional[Callable] = None):
        loss = None if closure is None else closure()
        for group in self.param_groups:
            # Get all hyperparas
            lr = group["lr"] 
            m = group["m"]
            v = group["v"]
            b1 = group["b1"]
            b2 = group["b2"]
            eps = group["eps"]
            lbd = group["lambda"]
            for p in group["params"]:
                """Iterate over each para in the group"""
                if p.grad is None:
                    continue

                # Get state associated with p.
                state = self.state[p] 
                t = state.get("t", 0) 
                m = state.get("m", torch.zeros_like(p))
                v = state.get("v", torch.zeros_like(p))
                
                # Get the gradient of loss with respect to p.
                grad = p.grad.data 

                # Update 1st & 2nd Moments Est.
                new_m = m*b1 + (1-b1)*grad
                new_v= v*b2 + (1-b2)*grad.pow(2)
                state["m"] = new_m
                state["v"] = new_v
                # Update lr
                new_lr = lr*(math.sqrt(1-(b2)**t)/(1-(b1)**t))
                group["lr"] = new_lr
                # Update weights and weights decay
                p.data -= new_lr * new_m/ (math.sqrt(v)+eps)  
                p.data -= new_lr*lbd*p.data
                # Increment iteration number.
                state["t"] = t + 1 
        return loss
    

# def optimizing(
#         model, iter_nums, data, loss_fn
#         lr=1e-3, 
#         b1=0.9, b2=0.95, 
#         lbd=0.9, 
#         eps=1e-8
#     ):
#     """Training model for `iter_nums` iterations"""
#     opt = AdamW(model.parameters(), lr, b1, b2, lbd, eps)
#     for i in range(1, iter_nums+1):
#         opt.zero_grad()
        


weights = torch.nn.Parameter(5 * torch.randn((10, 10)))
opt = SGD([weights], lr=1e3)
for t in range(10):
    opt.zero_grad() # Reset the gradients for all learnable parameters.
    loss = (weights**2).mean() # Compute a scalar loss value.
    print(loss.cpu().item())
    # print(opt.defaults["lr"])
    loss.backward() # Run backward pass
    opt.step()
```
